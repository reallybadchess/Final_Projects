```python
# import libraries
import pandas as pd
import numpy as np
from os import path
import matplotlib.pyplot as plt
```

```python
#reading csv into dataframe

df = pd.read_excel('office_supplies.xlsx')
```

```python
# quick peak at the df

df.head()
```

```python
# quick check on shape
df.shape
```

```python
# Searching for duplicates

dupes = df.duplicated()
df[dupes]
```

```python
# new df without dupes
df1 = df[~dupes]
```

```python
# checking shape once again
df1.shape
```

```python
# We will be missing an index entry so we need to reset index
df1.reset_index()
```

```python
# Dropping NaN columns
df1.drop(['Unnamed: 12','Unnamed: 13'], axis = 1, inplace = True)
df1.columns
```

```python
# checking for null values
df1.isna().sum()
```

#### Wow! no missing values. This is wonderful!

```python
df1.head()
```

```python
# making all columns lowercase 
df1.columns = df1.columns.str.lower()
```

```python
# replacing spaces with underscore
df1.columns = df1.columns.str.replace(' ','_')
```

```python
# Checking df
df1.head()
```

```python
# Saving to a xslx
df1.to_excel("office_supplies_cleaned.xlsx")  
```
